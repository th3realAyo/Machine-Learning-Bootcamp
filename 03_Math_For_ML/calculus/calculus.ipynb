```python
import numpy as np
import matplotlib.pyplot as plt
```

**Simple Line: y = 2x + 1**

```python
# Create a function
def lineFunc(x):
    return 2 * x + 1
```

```python
x = np.linspace(-3, 3, 100)
y = lineFunc(x)
```

```python
print(x[:5])
print(y[:5])
```

```python
# Find the slope of the x & y values

# m = y2-y1/x2-x1
plt.figure(figsize=(6,5))
plt.plot(x, y, color="blue")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Slope")
```

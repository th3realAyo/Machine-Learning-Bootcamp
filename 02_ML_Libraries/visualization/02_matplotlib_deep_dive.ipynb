```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Simple Line Plot** : A line chart connects data points with lines. It’s often called a “plot” or “line plot.”

- ***Use Cases:***
- Tracking trends over time (time series data).
- Showing continuous relationships between variables.
- Comparing multiple series.

#### This line of code is creating a scatter plot using the data in the `sales_df` DataFrame. It is specifying that the x-axis values should be taken from the index of the DataFrame (`sales_df.index`) and the y-axis values should be taken from the 'Sales' column. The title of the plot is set to 'Sales Scatter'.
**sales_df.plot(kind='scatter', x=sales_df.index, y='Sales', title='Sales Scatter')**

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)

print(len(x))
print(x[:5])


# `plt.plot(x, y)` is creating a line plot using the values in the array `x` as the x-coordinates and the values in the array `y` as the y-coordinates. 
# This function is used to visualize the relationship between the two arrays by connecting the data points with lines.
plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y = sin(x)")


plt.show()
```

**Customized Plot**

```python
plt.figure(figsize=(10,4))
plt.title("Customized Plot Chart")
plt.plot(x, y, color="red", linestyle = "--", marker = "o", label = "sin(x)")
plt.xlabel("X - values")
plt.xlabel("Y - values")
plt.legend()
plt.grid(True)
# plt.show()
```

**Bar Chart** : A bar chart uses rectangular bars to show the frequency or value of categories. Bars can be vertical or horizontal.
- ***Use Case***
- Comparing categorical data (e.g., sales by product, votes by candidate).
- Showing differences between groups.
- Highlighting trends across discrete categories.

```python
months = ["January", "February", "March", "April", "May", "June"]
sales = [120, 150, 170, 140, 200, 180]

plt.bar(months, sales, color = "purple")
plt.title("Sales Per Month - Bar Chart")
plt.xlabel("Months")
plt.ylabel("Sales Frequency")
# plt.grid(True)
plt.show()
```

**Horizontal Bar Chart**

```python
months = ["January", "February", "March", "April", "May", "June"]
sales = [120, 150, 170, 140, 200, 180]

plt.barh(months, sales, color="green")
plt.title("Sales - Horizontal Bar Chart")
plt.ylabel("Months")
plt.xlabel("Sales Frequency")
# plt.grid(True)
plt.show()
```

**Scatter Plot** : A scatter plot displays points based on two numerical variables, showing relationships, correlations, or clusters.

- ***Use Cases:***
- Analyzing correlation between variables.
- Detecting patterns or clusters.
- Spotting outliers in data.

```python
np.random.seed(42)
x = np.random.randn(50)
y = np.random.randn(50)

plt.scatter(x, y, color="red", alpha=0.6, s=5)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
# plt.colorbar()
plt.show()
```

**Histogram** : A histogram groups numerical data into bins and shows how often values occur. Unlike a bar chart, histograms deal with continuous data.

- ***Use Cases:***
- Understanding distribution of data.
- Detecting skewness, outliers, or patterns.
- Analyzing frequency of numeric ranges.

```python
value = np.random.randn(1000)

plt.hist(value, color = "skyblue", edgecolor = "black", bins=50)
plt.title("Data Distribution - Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import pandas as pd

data = {
    "Day": list(range(1, 31)),
    "Sales": [
        200, 220, 210, 205, 215, 230, 225, 240, 235, 250,
        245, 260, 255, 270, 265, 280, 275, 290, 285, 300,
        295, 310, 305, 320, 315, 330, 1000, 340, 50, 360
    ]
}

df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt

# Calculate IQR
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Split normal vs outliers
normal = df[(df['Sales'] >= lower_bound) & (df['Sales'] <= upper_bound)]
outliers = df[(df['Sales'] < lower_bound) | (df['Sales'] > upper_bound)]

# Plot
plt.figure(figsize=(10, 5))

plt.scatter(normal['Day'], normal['Sales'], color='blue', label='Normal')
plt.scatter(outliers['Day'], outliers['Sales'], color='red', label='Outliers')

plt.axhline(upper_bound, color='green', linestyle='--', label='Upper Bound')
plt.axhline(lower_bound, color='orange', linestyle='--', label='Lower Bound')

plt.xlabel('Day')
plt.ylabel('Sales')
plt.title('Scatter Plot with Outliers Highlighted')
plt.legend()

plt.show()
```

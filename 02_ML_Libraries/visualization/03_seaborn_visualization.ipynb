```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Load In-Built Seaborn Dataset**

```python
tips = sns.load_dataset("tips")
```

```python
tips.shape
```

```python
tips.head()
```

```python
# This code snippet is setting the global style and color palette for plotting using the Seaborn library in Python.

# set global style and color palette
sns.set_style("darkgrid")
sns.set_palette("viridis")
sns.set_theme()
```

**Histogram - Distribution of Total Bill**

```python
# sns.kdeplot(tips['total_bill'], fill=True, color="yell◘ow")

plt.figure(figsize=(7,4))
sns.histplot(tips["total_bill"], kde=True, color="black", edgecolor = "black")
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
```

**Box Plot - Detecting Outliers**

```python
plt.figure(figsize=(6,4))
sns.boxplot(x = "day", y="total_bill", data=tips, palette="pastel", hue="day", legend=False)
plt.title("Bill Distribution by Day")
plt.show()
```

**Count Plot - Number of Bill Per Day**

```python
plt.figure(figsize=(6,4))
sns.countplot(x = "day", data=tips, palette="pastel", hue="sex")
plt.title("Daily Bill Distribution by Gender")
plt.xlabel("Day")
plt.ylabel("Frequency")
plt.show()
```

**Scatter Plot  - Relationship**

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x="total_bill", y="tip", data=tips, size="size", palette="cool", hue="sex")
plt.title("Bill vs Tip Relationship")
plt.xlabel("Day")
plt.ylabel("Frequency")
plt.show()
```

**Pair Plot - Feature Relationship**

```python
sns.pairplot(tips, hue="sex", diag_kind="kde")
# plt.suptitle("Pair Wise Relationship")
plt.show()
```

**Heatmap - Feature Correlation**

```python
correlation_matrix = tips.corr(numeric_only=True)
plt.figure(figsize=(6,4))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

### **Multicolinearity Example**

```python
import pandas as pd
import numpy as np

np.random.seed(0)

# Create features
house_size = np.random.randint(500, 2000, 100)
number_of_rooms = house_size / 100 + \
    np.random.normal(0, 1, 100)  # highly correlated

# Target variable
price = house_size * 300 + np.random.randint(-10000, 10000, 100)

# DataFrame
df = pd.DataFrame({
    'house_size': house_size,
    'number_of_rooms': number_of_rooms,
    'price': price
})

print(df.head())
```

```python
print(df.corr())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

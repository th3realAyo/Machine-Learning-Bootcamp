**❓ What is Matplotlib**

Matplotlib is a Python library for creating visual charts and graphs. It allows you to plot data using lines, bars, histograms, scatter plots, and many other styles.

**⚙️ Why Matplotlib is Important in Machine Learning**

Matplotlib is the **foundation of data visualization** in Python.  
It helps us understand data patterns, relationships, and model performance through visual insights.

**🔍 Why Matplotlib is Essential:**
- 📊 **Data Exploration:** Visualize data distributions, correlations, and trends.
- 🤖 **Model Understanding:** Plot decision boundaries, training curves, feature importance.
- 📈 **Performance Evaluation:** Compare predictions vs actuals, loss curves, accuracy.
- 🧩 **Integration:** Works perfectly with NumPy and Pandas.

> 💡 *If NumPy handles numbers and Pandas handles data, Matplotlib helps you SEE them.*

**Simple Line Plot**

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y = sin(X)")
plt.show()
```

**Customized plot**

```python
plt.figure(figsize=(7, 4))
plt.plot(x, y, color="red", linestyle="--", marker="o", label="sin(x)")
plt.title("Customized Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid(True)
plt.show()
```

**Bar Chart**

```python
students = ['Alice', 'Bob', 'Charlie', 'David']
scores = [85, 92, 78, 88]

plt.bar(students, scores, color="orange")
plt.title("Students Scores - Bar Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```

```python
plt.barh(students, scores, color="orange")
plt.title("Horizontal Bar Chart")
plt.xlabel("score")
plt.ylabel("Students")
plt.show()
```

**Scatter Plot**

```python
np.random.seed(42)
```

```python
x = np.random.randn(50)
y = np.random.randn(50)

plt.scatter(x, y, color="purple", alpha=0.7)
plt.title("Scatter Plot")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

**Histogram**

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, color="skyblue", edgecolor="black")
plt.title("Data Distribution (Histogram)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

**Pie Chart**

```python
labels = ['Python', 'Java', 'C++', 'R']
sizes = [45, 30, 15, 10]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Programming Language Popularity")
plt.show()
```

**Multiple Plots (sub-plots)**

```python
x = np.linspace(0, 10, 100)

plt.figure(figsize=(10, 4))

plt.subplot(1,2,1)
plt.plot(x, np.sin(x), "r--")
plt.title("Sine")

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x), "b--")
plt.title("Cosine")

plt.suptitle("Sine and Cosine Functions")
plt.show()
```

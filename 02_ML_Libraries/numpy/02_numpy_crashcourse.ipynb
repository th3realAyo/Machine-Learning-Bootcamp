### Concatenate 2 or  more arrays

```python
a = np.arange(5, dtype=int)
b = np.arange(5, 10, dtype=int)
c = np.arange(10, 15, dtype=int)

print(np.concatenate((a, b, c))) # The variable must be converted to an iterable (tuple, list) else error
print(np.concatenate([a, b, c]))
```

### Reshape

```python
arr = np.arange(1, 10, dtype=int) #This generates 1,2,3,4,5,6,7,8,9 = 9 element
print(f"Original: {arr}")

# To reshape: product must equal total element

print(f"Reshaped 1: {arr.reshape(3, 3)}") # 3 x 3 = 9
print(f"Reshaped 2: {arr.reshape(9, 1)}") # 9 x 1 = 9
print(f"Reshaped 3: {arr.reshape(1, 9)}") # 1 x 9 = 9

# Creating a 3D Array
arr = np.arange(27, dtype=int)
print(f"3D : {arr.reshape(3, 3, 3)}")
print(f"3D : {arr.reshape(1,9,3)}")
print(f"3D : {arr.reshape(9,1,3)}")

# Indexing a reshaped array
target = arr.reshape(9,1,3)
print(target[5,0,1])
```

### Adding New Axis To An Array, Slicing, Indexing

```python
# Increase the dimension of an array
arr = np.arange(3)
print(arr[np.newaxis, :]) # Increase the column
print(arr[:, np.newaxis]) # Increase the row


# expand_dims adds a new axis (0 = row, 1 = column)
expanded = np.expand_dims(arr, axis=0)
print(expanded)
```

```python
# Slicing and Indexing
arr = np.arange(20, dtype=int)
# print(arr[0])
# print(arr[-1])
# print(arr[3:5])
# print(arr[::2])

# We can specify a condition for the return value
print(arr[arr % 2 == 0])
print(arr[arr % 2 == 1])
print(arr[arr < 3])
print(arr[arr > 10])

# Chained condition
print(arr[(arr % 2 == 0) & (arr > 10)])

# -------------------------------------
# -------------------------------------

arr = np.arange(20, dtype=int)
conditional = (arr % 2 == 0) | (arr == 15)
print(conditional)

# -------------------------------------
# -------------------------------------
# Non-Zero
matrix = np.arange(20, dtype=np.int16)
reshaped = np.reshape(matrix, (5, 4))
print(reshaped)
print((np.nonzero(reshaped % 2 == 0))) # Returns rows and columns that met requirement
print((np.nonzero(reshaped % 2 == 100))) # No condition met
```

```python
# Creating a Copy of an array using the .copy() method

arr = np.arange(10)
new_arr = arr[3:6]
new_arr[0] = 999

print(arr)
print(new_arr)

arr = np.arange(10)
new_arr = arr[3:6].copy()
new_arr[0] = 999

print(arr)
print(new_arr)
```

### UNIQUE ARRAY VALUES

```python
arr = np.array([1,1,1,2,3,4,4,5,66,4,3,3,2,4,5,5,3,2,2,4,56,4,3,22,2])

# 1️⃣ Unique values
print(np.unique(arr))  


# 2️⃣ Unique values + counts
print(np.unique(arr, return_counts=True))


# 3️⃣ Unique values + first index of each unique element
print(np.unique(arr, return_index=True))


# 4️⃣ Unique values + inverse indices
print(np.unique(arr, return_inverse=True))
```

**2. Function with parameters**

```python
def greet_user(name):
    print("Hello " + name + "!")
```

```python
greet_user(name="Sid")
```

```python
greet_user(name="John")
```

```python
greet_user("Sid")
```

**3. return a value**

```python
# using return
def square(n):
    return n*n

# save result in "result" variable
result = square(n=3)
print("Square is:", result)

# you can also print directly
print("Square of 10 is", square(10))
```

```python
# default paramaeters
def add(a=100, b=200):
    result = a+b
    print("Sum is:", result)
```

```python
add(a=10, b=20)
```

```python
add()
```

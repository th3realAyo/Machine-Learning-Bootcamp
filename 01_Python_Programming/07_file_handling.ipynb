**1. Opening & Reading a file**

```python
# open a file in read mode
f_path = r"C:\Users\AYO_AYO\Desktop\Bluechip_Machine_Learning\Udemy - ML\My_Note\01_THE_DIFFERENCE_BETWEEEN_AI_ML_DL.txt"
file = open(f_path, "r")

# read the content
content = file.read()

print(content)

file.close()
```

**2. Using with to read (auto-closing)**

```python
with open(f_path, "r") as file:
    content = file.read()
    print(content)
```

**3. Writing to a File**

```python
import os

f_path = r"C:\Users\AYO_AYO\Desktop\Bluechip_Machine_Learning\Udemy - ML\My_Note"

with open(os.path.join(f_path, "output.txt"), "w") as file:
    file.write("This is my first file")
```

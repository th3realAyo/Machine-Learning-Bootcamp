**1. Arithmetic Operators** - Used to perform mathematical oprations

```python
a = 10
b = 3

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a % b)
print(a ** b)
print(a // b)
```

**2. Comparison Operators** - Used to compare two values

```python
x = 5
y = 10

print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
```

**3. Logical Operators** - Used to combine multiple conditions

```python
x = 50
y = 10

print(x != y and x < y)
print(x != y or x < y)
```

```python
is_adult = True
has_id = False

print(is_adult and has_id) # print(True and False)
print(is_adult or has_id) # print(True or False)
print(not is_adult)
print(not has_id)
```

**4. Assignment Operators** - Assignment operators are used to assign values to variables:

- `=` : Assign
- `+=` : Add and assign
- `-=` : Subtract and assign
- `*=` : Multiply and assign
- `/=` : Divide and assign
- `%=` : Modulus and assign
- `**=` : Exponent and assign
- `//=` : Floor divide and assign

```python
number = 10
print(number)

number = number + 100

print(number)
```

```python
number = 10
print(number)

number += 100  # number = number + 100
print(number)
```

```python
# Assignment Operators Example
num = 7
num += 3  # num = num + 3
print(num)
num *= 2  # num = num * 2
print(num)
num -= 4  # num = num - 4
print(num)
num /= 2  # num = num / 2
print(num)
num %= 3  # num = num % 3
print(num)
num **= 2  # num = num ** 2
print(num)
num //= 2  # num = num // 2
print(num) 
```

**5. Membership Operators** - Membership operators test for membership in a sequence:

- `in` : True if value is found
- `not in` : True if value is not found

```python
a = [1, 2, 3, 4, 5]

print(3 in a)
print(10 in a)
```

```python
a = [1, 2, 3, 4, 5]

print(3 not in a)
print(10 not in a)
```

**6. Identity Operators** - Identity operators compare memory locations of two objects:

- `is` : True if both refer to the same object
- `is not` : True if they do not refer to the same object

```python
a = [1, 2, 3] # a and c both refer to same object [1, 2, 3]
b = [1, 2, 3]
c = a

print(a is b)
print(a is c)
print(a is not b)
```

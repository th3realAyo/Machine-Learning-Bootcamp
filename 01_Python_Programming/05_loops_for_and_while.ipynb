**2. While Loop**

```python
count = 1

while count <= 5:
    print("Count is", count)
    count += 1  # same as count = count + 1
```

**3. Loop Controls - break; continue**

```python
# break - stop a loop when a condition is reached

for i in range(0, 10):
    if i == 3:
        break
    print(i)
```

```python
# continue - skip an iteration
for i in range(0, 10):
    if i == 3:
        continue
    print(i)
```

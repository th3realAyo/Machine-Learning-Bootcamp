## 1. Define a Basic Class

Let's start by defining a simple class using the `class` keyword.

```python
# define a simple class
class Dog:
    pass
```

## 2. Create Class Instances

Now, let's create objects (instances) from our class.

```python
# creating an instance of dog class
my_dog = Dog()
```

```python
print(type(my_dog))
```

## 3. Add Methods to a Class

Methods are functions defined inside a class. Let's add a method to our Dog class.

```python
# add method to Dog class

class Dog:
    def bark(self):
        print("Woof!")
```

```python
my_dog = Dog()
my_dog.bark()
```

```python
dog_1 = Dog()
dog_1.bark()
```

## 4. Use Class Attributes

Attributes are variables that belong to a class. Let's add and use attributes in our Dog class.

```python
# add attributes to Dog class
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def bark(self):
        print("Woof!")
```

```python
my_dog = Dog(name="Buddy", age=3)
print(my_dog.name)
print(my_dog.age)
my_dog.bark()
```

```python
dog_2 = Dog(name="Leo", age=4)
print(dog_2.name)
print(dog_2.age)
dog_2.bark()
```

## 5. Demonstrate Inheritance

Inheritance allows one class to use the properties and methods of another class. Let's create a subclass that inherits from Dog.

```python
# add attributes to Dog class
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def bark(self):
        print("Woof!")
```

```python
# create a subclass that inherits from Dog
class Puppy(Dog):
    def play(self):
        print(self.name, "is playful")
```

```python
my_puppy = Puppy(name="Max", age=1)
```

```python
print(my_puppy.name)
print(my_puppy.age)
my_puppy.bark()
my_puppy.play()
```

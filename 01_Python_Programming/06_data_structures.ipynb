**List - Ordered & Changeable (Mutable)**

```python
# creating a list
fruits = ["apple", "banana", "mango"]
print(fruits)
```

```python
# Accessing items using index (starts at 0)
print(fruits[0])  # apple
print(fruits[1])  # banana
```

```python
# change an element
fruits[2] = "orange"
print(fruits)
```

```python
a = [1, 2, 3.5, False, "word"]
print(a)
```

```python
print(type(a))
```

```python
# add an item to a list
a.append("apple")
print(a)
```

```python
# remove an element from a list
a.remove("word")
print(a)
```

```python
# find teh number of elements in a list
print(len(a))
```

```python
# loop through a list
fruits = ["apple", "banana", "mango"]

for fruit in fruits:
    print(fruit)
```

**Tuple - Ordered & Unchangeable (Immutable)**

```python
# tuple of colors
colors = ("red", "green", "blue")
print(colors)
print(type(colors))
```

```python
print(colors[0])
print(colors[2])
```

```python
# Trying to change an item (this will cause an error)
# colors[1] = "yellow"  # ❌ Error: 'tuple' object does not support item assignment
```

```python
for i in colors:
    print(i)
```

```python
a = [1, 2 ,3, 4, 5]

b = tuple(a)

print(type(a))
print(type(b))

print(a)
print(b)
```

**Set - Unordered & Unique**

```python
# set of student names
students = {"Alice", "John", "Sid"}
print(students)
print(type(students))
```

```python
students.add("David")
print(students)
```

```python
students.remove("Alice")
print(students)
```

```python
# Loop through set
for student in students:
    print(student)
```

```python
list_1 = [1, 1, 2, 3, 3, 5]
print(list_1)
```

```python
set_1 = {1, 1, 2, 3, 3, 5}
print(set_1)
```

```python
list_2 = [10, 9, 9, 8, 5, 1, 1, 2, 4, 8]
print(list_2)

list_2 = list(set(list_2))
print(type(list_2))
print(list_2)
```

**Dictionary - Key-value pair data**

```python
# dictionary with a person details
person = {
    "name": "Alice",
    "age": 25,
    "is_student": True
}
```

```python
person["age"]
```

```python
person["name"]
```

```python
person["age"] = 26
```

```python
print(person)
```

```python
print(type(person))
```

```python
person["city"] = "London"
```

```python
print(person)
```

```python
# loop through a dict in for loop
for key, value in person.items():
    print(key, "->", value)
```

### ✅ Python Data Structures Comparison

| **Type**      | **Ordered**     | **Changeable** | **Allows Duplicates** | **Indexable**       |
|---------------|------------------|----------------|------------------------|----------------------|
| **List**       | ✅ Yes            | ✅ Yes         | ✅ Yes                 | ✅ Yes               |
| **Tuple**      | ✅ Yes            | ❌ No          | ✅ Yes                 | ✅ Yes               |
| **Set**        | ❌ No             | ✅ Yes         | ❌ No                  | ❌ No                |
| **Dictionary** | ✅ Yes *(by key)* | ✅ Yes         | ❌ No *(keys only)*    | ✅ Yes *(by key)*    |
